```
samp = sc.textFile('/FileStore/tables/sample_2.txt')
samp.collect()
```

```
#Getting the lenght of each word with functions
def lenth(x):
  nl = x.split(' ')
  nl2 = []
  for i in nl:
    nl2.append(len(i))
  return nl2

rdds = samp.map(lenth)
rdds.collect()
```

```
#Getting the lenght of each word with lambda

lm = samp.map(lambda x: [len(s) for s in  x.split(' ')])
lm.collect()
```

```
# Saving into a text file defaults the partitions to 2, i.e. save the content into 2 partitions
# the Repartition function allows you to state how many partitions you want to split content of the save text file to.

from pyspark import SparkConf, SparkContext 
conf = SparkConf().setAppName("FlatMap") 
sc = SparkContext.getOrCreate(conf=conf)


wodi_part = sc.textFile('/FileStore/tables/sample_3.txt')
wodi_part_a = wodi_part.repartition(5)
wodi_part_b = wodi_part_a.map(lambda x:(x,1))
```

```
# Remove Header
stud = sc.textFile('/FileStore/tables/StudentData.csv')
head = stud.first()
stud1 = stud.filter(lambda x: x != head)
```

```
# Get number of students - Use count ()
stud1.count()
```

```
stud2 = stud1.map(lambda x:x.split(','))
stud2.collect()
```

```
# Show the total marks achieved by Female and Male students 
stud3 = stud2.map(lambda x:(x[1], int(x[5])))
stud3.reduceByKey(lambda x,y: x+y).collect()
```

```
# Show the total number of students that have passed and failed. 50+ marks are required to pass the course.

no_Passed = stud2.filter(lambda x:int(x[5]) > 50).count()
no_Failed= stud2.filter(lambda x:int(x[5]) <= 50).count()

print('Student who failed =', no_Failed)
print('Student who Passed =', no_Passed)
```

```
# Show the total number of students enrolled per course
enrolled = stud2.map(lambda x:(x[3], 1))
enrolled.reduceByKey(lambda x,y:x+y).collect()
```

```
# Show the total marks that students have achieved per course
TotalMarks = stud2.map(lambda x:(x[3], int(x[5])))
TotalMarks.reduceByKey(lambda x,y:x+y).collect()
```

```
# Show the average marks that students have achieved per course
Average_mark = stud2.map(lambda x:(x[3],(int(x[5]), 1)))
Average_mark2 = Average_mark.reduceByKey(lambda x,y:x+y)

new = Average_mark.reduceByKey(lambda x,y: (x[0]+y[0], x[1] + y[1]))
new.collect()
```

```
# new.map(lambda x:(x[0], x[1][0]/x[1][1])).collect()
new.mapValues(lambda x:(x[0]/x[1])).collect() # mapValues, take the key, and uses as a reference for any calculations on the secound values (similar to reduceByKey())
```

```
# Get the Minimum and Maximun marks per course
Min_marks = stud2.map(lambda x:(x[3],int(x[5])))
Min_marks.collect()
```

```
Max = Min_marks.reduceByKey(lambda x,y: x if x > y else y)
Min = Min_marks.reduceByKey(lambda x,y: x if x < y else y)
```

```
Max.collect()
Min.collect()
```
